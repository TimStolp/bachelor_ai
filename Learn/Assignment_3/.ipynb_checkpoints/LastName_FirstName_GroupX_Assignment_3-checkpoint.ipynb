# K-means

For this weeks assignment, the setup will be somewhat different from the previous weeks. We will jump ahead a little to an algorithm that hasn't been covered yet: *K-means*. K-means is a classic example of *unsupervised learning*, where we do not have a target output we are trying to model, but instead are just trying to extract a pattern from the whole dataset. So hopefully, together with the previous 2 programming assignments, this should round out your perspective of concrete implementations of the most common types of *machine learning* algorithms.

The goal of *K-means* is to find $k$ clusters, or groups, within a dataset. Start by reading the description of the algorithm given in section *7.3* of Alpaydin. The resulting pseudo-code for the algorithm to find $k$ clusters, is shown there in figure *7.3*:

* Initialize $m_i$, $i$ = $1$, ... , $k$, for example, to $k$ random $x^t$
* Repeat
    * For all $x^t \in X$
        * $b_i^t \leftarrow \left\{\begin{array}{ll} 1 & if\ \Vert x^t − m_i \Vert \ =\ min_j\ \Vert x^t − m_j \Vert\\ 
            0 & otherwise \\ \end{array}\right.$
    * For all $m_i$, $i$ = $1$, ... , $k$
        * $m_i \leftarrow \sum_t b_i^t x^t / \sum_t b_i^t$
* Until $m_i$ converge

There are a couple of videos by Andrew Ng on the topic that might also be helpful to understand what the algorithm does. The notation here is a little different, but the ideas are exactly the same:

## Showing the results [3 pt]

To show your code works, run the algorithm on the Iris dataset using only the last 2 variables of the data. Use a value of 3 for $k$ and plot the means with their assignments. Start by running the algorithm step by step and plotting for each step. Start with the random means and then show at least 2 more steps, to show the means moving as the algorithm iterates. Finally, show the resulting plot where the means have converged and the algorithm has stopped.


```python
import math
import matplotlib.pylab as plt
import numpy as np
import random

from sklearn import datasets

data = datasets.load_iris().data[:, 2:]

def initialize_means(data, k):
    return [random.choice(data) for i in range(k)]

def compute_distance(a,b):
    return np.linalg.norm(a-b)

def assign_points(data, means):
    matrix_b = []
    for point in data:
        temp_list = []
        for mean in means:
            temp_list.append(compute_distance(point,mean))
        matrix_b.append(np.append(point,temp_list.index(min(temp_list))))
    return np.array(matrix_b)
        
def calculate_new_means(matrix_b,k):
    new_means_list = []
    for i in range(k):
        if matrix_b[matrix_b[:,2] == i].size > 0:
            new_means_list.append(matrix_b[matrix_b[:,2] == i][:,:2].mean(axis=0))
        else:
            new_means_list.append(initialize_means(data,1))
    return np.array(new_means_list)
    
def plot_clusters(matrix_b, means):
    for i in range(len(means)):
        plt.plot(matrix_b[matrix_b[:,2] == i][:,0],matrix_b[matrix_b[:,2] == i][:,1],"o",alpha=0.5)
        plt.plot(means[i][0],means[i][1],"kx",mew=2.5,ms=10)
    return 0

def check_converged(means,new_means):
    if all(x in new_means for x in means):
        return 1
    else:
        return 0

def k_means(data,k,iterations):
    means = initialize_means(data,k)
    for i in range(iterations):
        matrix_b = assign_points(data,means)
        plot_clusters(matrix_b,means)
        plt.show()
        new_means = calculate_new_means(matrix_b,k)
        if check_converged(means,new_means):
            break
        else:
            means = new_means
    return 0

k_means(data,3,100)
```

```python
# I define a new k_means here which does not plot each step because the cost function calls k_means many times,
# and it returns matrix_b and the mean vectors to be able to use them in the cost function.
def k_means_no_plot(data,k,iterations):
    means = initialize_means(data,k)
    for i in range(iterations):
        matrix_b = assign_points(data,means)
        new_means = calculate_new_means(matrix_b,k)
        if check_converged(means,new_means):
            break
        else:
            means = new_means
    return matrix_b, means


def cost_function(matrix_b,means):
        return 1/len(matrix_b)*sum([compute_distance(matrix_b[i][:1],means[int(matrix_b[i][2])])**2 for i in range(len(matrix_b))])
      
def cost_function_average(matrix_b,means,iterations):
    accumulator = 0
    for i in range(iterations):
        accumulator += cost_function(matrix_b,means)
    return 1/iterations*accumulator
        
def cost_function_plot(data):
        cost_list = []
        for i in range(1,11):
            matrix_b, means = k_means_no_plot(data,i,100)
            cost_list.append(cost_function_average(matrix_b,means,50))
        plt.plot(np.linspace(1,10,num=10),cost_list)
        plt.show()
        return 0

cost_function_plot(data)
```
