### Question 1(c) (5pts)
Give three points (stimuli) that you can use to determine whether smize is separable from height. (3pts)

> s2h1, s1h1, s1h2

Explain how you will use these to indicate whether or not the dimensions of smize and height are separable. (2pts)

> If they were separable, (s2h1, s1h1) + (s1h1, s1h2) = (s2h1, s1h2), through Taxicab distance. 

### Question 1(d) (5pts)
Using the triple of points you gave in question 1(c), determine whether smize and height are integral or separable.

```python
from math import sqrt

path1 = values[('s2h1', 's1h1')]
path2 = values[('s1h1', 's1h2')]
finalpath = values[('s2h1', 's1h2')]

taxicab = path1 + path2
euclidean = sqrt(path1**2 + path2**2)

if abs(euclidean - finalpath) > abs(taxicab - finalpath):
    print('Separable')
else:
    print('Integral')
```

```python
dist_list = []
cab_list = []
eucl_list = []
for key in values:
    a_1 = int(key[0][1])
    a_2 = int(key[0][3])
    b_1 = int(key[1][1])
    b_2 = int(key[1][3])
    dist_list.append(values[key])
    cab_list.append(abs(a_1 - b_1) + abs(a_2 - b_2))
    eucl_list.append(sqrt((a_1 - b_1)**2 + (a_2 - b_2)**2))

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(eucl_list, dist_list, 'o')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('Euclidean')

plt.subplot(122)
plt.plot(cab_list, dist_list, 'o')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('City Block')

plt.show()
```

### Question 1f) (10 pts)

```python
from scipy import stats

slope_eucl, intercept_eucl, r_value_eucl, p_value_eucl, std_err_eucl = stats.linregress(eucl_list, dist_list)
slope_cab, intercept_cab, r_value_cab, p_value_cab, std_err_cab = stats.linregress(cab_list, dist_list)

eucl_line_values = [slope_eucl * i + intercept_eucl for i in eucl_list]
cab_line_values = [slope_cab * i + intercept_cab for i in cab_list]


plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(eucl_list, dist_list, 'o')
plt.plot(eucl_list, eucl_line_values, 'y')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('Euclidean')

plt.subplot(122)
plt.plot(cab_list, dist_list, 'o')
plt.plot(cab_list, cab_line_values, 'y')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('City Block')

plt.show()
```

### Question 2c) (20pts)
The Haussdorff distance was defined in Lecture 1-2 on slides 27-28 using the following 3 equations:

 $$ d_S(x, Y) = \inf\{d(x, y) | y \in Y\} $$
 
 where $d(x, y)$ is the distance between two points.

$$ d_{SS}(X, Y) = \sup\{d_S(x,Y) | x \in X\} $$

$$ d_H(X, Y) = \max\{d_SS(X, Y), d_SS(Y, X)\} $$ 

Write a function to calculate the Hausdorff distance between two sets, and calculate the distance between the set of emotions belonging to 'joy' and the other 7 sets of emotions, assuming a Euclidean distance on the VAD space. Print the distances. Which is 'joy' nearest to (apart from 'joy' itself)? Which is it furthest from?

```python
# For some words there was no vector data so we skipped these words when calculating the distance.

def hausdorff(e1, e2):
    distance_list = []
    # loop through sub emotions of emotion e1
    for sub_e in emotion_dict[e1]:
        # Get vector data if available
        if sub_e in emotion_vector_dict:
            sub_e_vec = emotion_vector_dict[sub_e]
        else:
            continue
        temp_distance_list = []
        # Loop through sub emotions of emotion e2
        for sub_e2 in emotion_dict[e2]:
            if sub_e2 in emotion_vector_dict:
                sub_e2_vec = emotion_vector_dict[sub_e2]
            else:
                continue
            # Calculate minimal distance between each sub emotion of e1 to all sub emotions of e2
            temp_distance_list.append(sqrt(sum([(x1 - x2)**2 for x1, x2 in zip(sub_e_vec, sub_e2_vec)])))
        # Choose minimal distance for each sub emotion of e1 to all emotions of e2
        distance_list.append(min(temp_distance_list))
    # Return maximum distance of all minimal distances
    return max(distance_list)


for key in emotion_dict:
    if key == 'joy':
        continue
    print(f"Distance from joy to {key} = {hausdorff('joy', key)}")
```

Joy is closest to fear and furthest away from disgust.
