In Lecture 1-1, slide 42 onwards, we described how we can decide whether dimensions are integral or separable. We do this by looking at the similarities that participants assign to pairs of points in the space.

Scientists have recently discovered a new sensory modality 'smision'. They want to test whether 'smize' is separable from height, and do this by preparing an experiment in the smize-height feature space.

They prepare 8 stimuli which can have values 0, 1, 2 on the height axis and 0, 1, 2 on the smize axis. The stimuli take all combinations of values except (1, 1). I.e., the 8 stimuli have values: 

[(0, 0), (0, 1), (0, 2), (1, 0), (1, 2), (2, 0), (2, 1), (2, 2)],

[s1h1, s1h2, s1h3, s2h1, s2h3, s3h1, s3h2, s3h3] respectively.


### Question 1(a) (5 pts)
Plot the stimuli in the smize-height feature space with smize on the x-axis and height on the y-axis. Label the axes and the individual points.

```python
#The stimuli are the pairs of values described in the question.
tuples = [(0, 0), (0, 1), (0, 2), (1, 0), (1, 2), (2, 0), (2, 1), (2, 2)]
names = ['s1h1', 's1h2', 's1h3', 's2h1', 's2h3', 's3h1', 's3h2', 's3h3']
x = [x[0] for x in tuples]
y = [x[1] for x in tuples]
fig, ax = plt.subplots()
ax.scatter(x, y)
for i, name in enumerate(names):
    ax.annotate(name, (x[i], y[i]))
plt.xlabel('x')
plt.show()
```

### Question 1(e) (10 pts)
The stimuli are a certain number of similarity steps apart in the feature space, e.g s1h1 one similarity step from s1h2, and 2 similarity steps from s1h3. So the distance between s1h1 and s1h2 in the feature space is 1, and the distance between s1h1 and s1h3 in the feature space is 2. The distance between s1h2 and s2h1 depends on whether the Euclidean metric or the cityblock metric should be used.

The dissimilarity ratings should be some function of the distance in the feature space. Make a figure with two subplots that plots dissimilarity as a function of distance in the feature space, firstly with distance calculated according to the Euclidean metric, and the secondly with distance calculated according to the city block metric.

You should get the following plots. Remember to add axis labels and titles.



```python
cab_dist_list = []
euc_dist_list = []
diss_list = []
for key in values:
    a_1 = int(key[0][1])
    a_2 = int(key[0][3])
    b_1 = int(key[1][1])
    b_2 = int(key[1][3])
    diss_list.append(values[key])
    cab_dist_list.append(abs(a_1 - b_1) + abs(a_2 - b_2))
    euc_dist_list.append(math.sqrt(abs(a_1 - b_1)**2 + abs(a_2 - b_2)**2))

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(euc_dist_list, diss_list, 'o')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('Euclidean')

plt.subplot(122)
plt.plot(cab_dist_list, diss_list, 'o')
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('City Block')


plt.show()
```

### Question 1f) (10 pts)
The plots should indicate that there is a linear relationship between distance in feature space and dissimilarity. Use linear regression to fit a line to the data. Plot the line on the previous axes and print the values of the coefficients.

```python
from scipy import stats
import numpy as np

plt.figure(figsize=(12,5))


plt.subplot(121)
plt.plot(euc_dist_list, diss_list, 'o')
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(euc_dist_list,diss_list)
plt.plot(euc_dist_list, [slope1 * x + intercept1 for x in euc_dist_list])
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('Euclidean')

plt.subplot(122)
plt.plot(cab_dist_list, diss_list, 'o')
slope2, intercept2, r_value2, p_value2, std_err2 = stats.linregress(cab_dist_list,diss_list)
plt.plot(cab_dist_list, [slope2 * x + intercept2 for x in cab_dist_list])
plt.xlabel('Distance')
plt.ylabel('Dissimilarity')
plt.title('City Block')


plt.show()
```
