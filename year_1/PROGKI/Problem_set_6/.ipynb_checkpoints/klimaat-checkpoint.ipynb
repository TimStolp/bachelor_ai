```python
# Max and Min temp with date

##### finds the rows of the max and min temperature
##### seperates the date number of a row into day, month, and year
##### prints temperature values with date written as DD MM YYYY with months as abbreviation of 3 letters
```

```python
import pandas as pd
import datetime

data = pd.read_csv('klimaat.csv')
max_row = data['TX'].idxmax()
min_row = data['TX'].idxmin()

def getdate(row):
    date = str(data.at[row, 'DATE'])
    day = int(date[6:8])
    month = int(date[4:6])
    year = int(date[0:4])
    return datetime.datetime(year, month, day).strftime('%d %b %Y')

print('De maximale temperatuur was {} graden op {}'.format(data.at[max_row, 'TX']/10, getdate(max_row)))
print('De minimale temperatuur was {} graden op {}'.format(data.at[min_row, 'TX']/10, getdate(min_row)))
```

```python
# Longest freezing span of days

##### checks when temp is below 0
##### gets the sum of days
##### gets the row of the last day that it froze furing the longest freezing period
```

```python
check = data['TX'] < 0
sum_of_days = check.ne(check.shift()).cumsum()

date_row = sum_of_days[sum_of_days == sum_of_days[check].value_counts().idxmax()].index[-1]

print('De langste vriesperiode is {} dagen en eindigde op {}.'.format(sum_of_days[check].value_counts().max(), getdate(date_row)))
```

```python
# graph of amount of summery days each year

##### make a check for when temp is above 25 degrees celcius
##### create function to get the year from a row
##### filter the dataframe with the check and group it by year and get the size of each group
##### size of group is the amount of days in that year that were above 25 degrees celcius
##### plot graph with years on x-axis and amount of summery days on y-axis
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

summery_check = data['TX'] >= 250

def get_year(row):
    return str(data.at[row, 'DATE'])[0:4]

counts = data[summery_check].groupby(get_year).size()

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 115
plt.rcParams["figure.figsize"] = fig_size
plt.plot(counts.index, counts)
plt.xlabel('years')
plt.ylabel('summery days')
plt.title('amount of summery days that each year had')
plt.show()
```
